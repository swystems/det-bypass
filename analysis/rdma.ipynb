```python
def scatter_latencies(filename):
    with open(filename, "r") as f:
        data = list(map(float, f.readlines()))
    yvalues = [val for val in data] #if val < 1e4]
    xvalues = range(len(yvalues))
    fig, axs = plt.subplots(1,1)
    axs.scatter(xvalues, yvalues, s=2)
```

```python
scatter_latencies("./data/rc/latencies.txt")
```
